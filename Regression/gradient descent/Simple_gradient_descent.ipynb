```
import numpy as np
import math
import pandas as pd
```

```
df=pd.read_csv("/content/drive/MyDrive/ml/test_scores.csv")
X=df["math"]
Y=df["cs"]
```

```
def grad(a,b):
  x=np.array(a.tolist())
  y=np.array(b.tolist())
  n=x.shape[0]
  lr=0.0002
  iter=1000
  m=0
  b=0
  errp=0
  for i in range(iter):
    y_pred=m*x+b
    err=(sum([val**2 for val in (y-y_pred)]))/n
    mb=-(2/n)*(sum(x*(y-y_pred)))
    bb=-(2/n)*(sum(y-y_pred))
    m=m-lr*mb
    b=b-lr*bb
    if math.isclose(err, errp, rel_tol=1e-20):
      break
    errp=err
    print(f"m= {m} b= {b} the error after {i} th iteration is {err}")
  return(m,b)
ma,ba=grad(X,Y)
```
