```
from word2number import w2n
import numpy as np
import math
e=0
for i in range(df["experience"].size):
   try:
     df["experience"][i]=w2n.word_to_num(str(df["experience"][i]))
   except:
     e=e+1
df2=df[df["experience"].isnull()==False]
mn=math.floor(df2["experience"].mean())
df.loc[(df["experience"].isnull()),'experience']=mn
df
```

```
df2=df[df["test_score(out of 10)"].isnull()==False]
mn=math.floor(df2["test_score(out of 10)"].mean())
df.loc[(df["test_score(out of 10)"].isnull()),'test_score(out of 10)']=mn
df
```
