```
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression as lr
```

```
df=pd.read_csv("/content/drive/MyDrive/ml/carprices (1).csv")
df
```

```
df2=pd.get_dummies(df["Car Model"])
df=pd.concat([df,df2],axis=1)
df.drop(["Mercedez Benz C class","Car Model"], axis=1, inplace=True)
df
```

```
Y=df["Sell Price($)"]
X=df.drop(["Sell Price($)"],axis=1)
X,Y
```

```
mod=lr().fit(X,Y)
mod.coef_
```

```
mod.predict([[33340,2,0,1]])
```

```
from sklearn.externals import joblib as jb
jb.dump(mod, "OHE_carprice")
```

```
mod2=jb.load("OHE_carprice")
```
